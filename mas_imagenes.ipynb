```python
import pandas as pd
import shutil
```

```python
df= pd.read_csv("data/facial_expresions/legend.csv")
```

```python
def path_create(x):
    return f"data/facial_expresions/images/{x}"
```

```python
df.head()
```

```python
df["path"][0]
```

```python
df["path"] = df["image"].apply(path_create)
```

```python
df["emotion"] = df["emotion"].apply(lambda x: x.lower())
```

```python
df["emotion"].unique()
```

```python
def copy_imgs(row):
    try:
        emocion = row["emotion"]
        image_name = row["image"]

        imagen_path = row["path"]
        new_path= f"data/facial_expresions/{emocion}/{image_name}"

        shutil.move(imagen_path,new_path)
    except:
        print("fallo en , ", image_name )
```

```python
df.apply(copy_imgs,axis=1)
```
